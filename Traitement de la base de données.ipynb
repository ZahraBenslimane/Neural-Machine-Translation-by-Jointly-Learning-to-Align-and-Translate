```python
def transform_xml (file1_eng , file2_fr ,write_file = 'mini_base_donnees.txt'  ,read_mode = 'w') :
    tree = ET.parse(file1_eng)
    tree2 = ET.parse(file2_fr)
    root = tree.getroot()
    root2 = tree2.getroot()
    f = open(write_file , read_mode)
    for doc in root.iter('doc') : 
        for doc2 in root2.iter('doc') : 
            if doc.attrib == doc2.attrib : 
                for seg in doc :
                    if seg.tag == 'seg' : 
                        for seg2 in doc2 : 
                            if seg2.tag == 'seg' : 
                                if seg.attrib == seg2.attrib : 
                                    print('{}\t{}'.format(seg.text , seg2.text))
                                    f.write(seg.text)
                                    f.write('\t')
                                    f.write(seg2.text)
                                    f.write('\n')
    f.close()
    return
    
```

```python
transform_xml ('IWSLT17.TED.dev2010.en-fr.en.xml' , 'IWSLT17.TED.dev2010.en-fr.fr.xml' ,write_file = 'mini_base_donnees_3.txt'  ,read_mode = 'w')
```

```python
def concat_2_text (file1_eng , file2_fr ,write_file = 'base_donnees_complete.txt'  ,read_mode = 'w' ) :
    f = open(write_file , read_mode)
    with open(file1_eng) as f1, open(file2_fr) as f2:
        for x, y in zip(f1, f2):
            f.write(x.strip())
            f.write('\t')
            f.write(y.strip())
            f.write('\n')
    f.close()
    return
```
